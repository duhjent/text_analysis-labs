# Imports

```python
import pandas as pd
import jsonlines
import json
from tqdm import tqdm
```

```python
with jsonlines.open('./data/books.jsonlines') as reader:
    for obj in reader:
        print(json.dumps(obj, ensure_ascii=False, indent=2))
        break
```

```python
with jsonlines.open('./data/reviews.jsonlines') as reader:
    for obj in reader:
        print(json.dumps(obj, ensure_ascii=False, indent=2))
        break
```

```python
df = []
with jsonlines.open('./data/reviews.jsonlines') as reader:
    for obj in reader:
        df.append({'title': obj['reviewTitle'], 'review': obj['review'], 'rating': obj['rating']})

df = pd.DataFrame(df)
df
```
