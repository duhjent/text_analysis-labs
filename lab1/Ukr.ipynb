# Imports

```python
from mitie import tokenize, named_entity_extractor, to_default_str_type
from collections import Counter
import pandas as pd
```

# Read the file

```python
with open('./ukr_text.txt') as file:
    f = file.read().lower()
```

# Tokenize

```python
tokens = tokenize(f)
```

```python
[to_default_str_type(t) for t in tokens]
```

# Tokens counter

```python
cnt = Counter(to_default_str_type(t) for t in tokens)
```

```python
cnt.most_common(25)
```

```python
cnt
```

# Stop-words

```python
stop_words = !cat stopwords_ua.txt

stop_words = set(stop_words)
```

```python
[to_default_str_type(t) for t in tokens if to_default_str_type(t) not in stop_words]
```

```python
from string import punctuation

tokens = [t for t in tokens if to_default_str_type(t) not in stop_words and to_default_str_type(t) not in punctuation]
```

```python
[to_default_str_type(t) for t in tokens]
```

# Stemming

```python
from uk_stemmer import UkStemmer

stemmer = UkStemmer()

tokens_str = [to_default_str_type(t) for t in tokens]

stems = [stemmer.stem_word(t) for t in tokens_str]

stems
```

```python
cnt = Counter(stems)

cnt.most_common(25)
```

```python
cnt
```
