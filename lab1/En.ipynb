# Imports

```python
from mitie import tokenize, named_entity_extractor, to_default_str_type
from collections import Counter
import pandas as pd
import nltk
import pandas as pd
from string import punctuation
```

# Read the file

```python
with open('./en_text.txt') as file:
    f = file.read().lower()
```

# Tokenize

```python
tokens = tokenize(f)
```

```python
[to_default_str_type(t) for t in tokens]
```

# Tokens counter

```python
cnt = Counter(to_default_str_type(t) for t in tokens)

cnt.most_common(25)
```

```python
cnt
```

# Stop-words

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
tokens = [t for t in tokens if to_default_str_type(t) not in stopwords and to_default_str_type(t) not in punctuation]
```

```python
[to_default_str_type(t) for t in tokens]
```

# Stemming

## Snowball

```python
stemmer = nltk.stem.snowball.SnowballStemmer('english')

[stemmer.stem(to_default_str_type(t)) for t in tokens if to_default_str_type(t) not in stopwords]
```

## Lancaster

```python
stemmer = nltk.stem.lancaster.LancasterStemmer()

[stemmer.stem(to_default_str_type(t)) for t in tokens if to_default_str_type(t) not in stopwords]
```

# Lemmatization

```python
lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()

[lemmatizer.lemmatize(to_default_str_type(t)) for t in tokens]
```

```python
cnt_lemma = Counter([lemmatizer.lemmatize(to_default_str_type(t)) for t in tokens if to_default_str_type(t) not in stopwords])
```

```python
cnt_lemma.most_common(25)
```

# POS-tagging

```python
tokens_str = [to_default_str_type(t) for t in tokens]

pos_df = pd.DataFrame(nltk.pos_tag(tokens_str), columns=['char', 'POS'])

pos_df
```
